```python
#start menu
def start():
        fno=int(input("""enter the function no. you want to perform
                        1.Login
                        2.Signup
                        3.Exit
                        
        """))
        if fno==1:
            login()
        else:
            if fno==2:
                signup()
            else:
                if fno==3:
                    print("bye bye")
                else:
                    print("invalid function no.")


#Login Function
def login(): 
    acc=int(input("enter the account number"))
    if acc in Bank:
        num=acc
        user=input("enter the username")
        passwd=input("enter the password")
        if user == Bank[acc]['Username']:
            if passwd == Bank[acc]['Password']:
                 for values in Bank[acc].items():    
                        print(*values,sep='\t =\t')
                 loginmenu(num)
                 repeat(num)
            else:
                print("invalid password")
        else:
            print("invalid username")
    else:
        print("invalid account number")
# login menu
def loginmenu(accno):
    no=accno
    lno=int(input("""enter the function no. you want to perform
                  1.Debit
                  2.Credit
                  3.Check
                  4.Logout
    
    """))
    if lno==1:
            debit(no)
    else:
        if lno==2:
            credit(no)
        else:
            if lno==3:
                check(no)
            else:
                if lno==4:
                    logout()
                else:
                    print("invalid function no.")

# Debit function
def debit(acc):
    no=acc
    damt=int(input("enter the amount you want to debit"))
    if damt>Bank[acc]['Balance']:
        print("no money")
    else:
        Bank[acc]['Balance']=Bank[acc]['Balance']-damt  
    repeat(no)

#credit function
def credit(acc):
    no=acc
    camt=int(input("enter the amount you want to credit"))
    if camt:
         Bank[acc]['Balance']=Bank[acc]['Balance']+camt
    else:
        print("Invalid amount")
    repeat(no)

#check function
def check(acc):
    no=acc
    print(f"your balance is { Bank[acc]['Balance']}")
    repeat(no)

#logout function
def logout():
    start()

#repeat function
def repeat(no):
    num=no
    opt=int(input("Press 1 for exit or press 2 to go back"))
    if opt==1:
        start()
    else:
        loginmenu(num)

#Signup
def signup():
    maxkey=max(Bank.keys())
    maxi=maxkey+1
    nme=input("Name of the account holder")
    uname=input("Username for account")
    paswrd=input("Password for the account")
    bala=int(input("Money you want to add to the account"))
    newaccount(maxi,nme,uname,paswrd,bala)
    opt=int(input("Press 1 for Login page or 2 for exit"))
    if opt==1:
        login()
    else:
        if opt==2:
            print("bye bye")
        else:
            print("invalid option")
                  
#new account 
def newaccount(acc,Name,Username,Password,Balance):
    Bank[acc]={'Name':Name,
               'Username':Username,
               'Password':Password,
               'Balance':Balance}
    print("Your account created successfully")
    print(f"Your account number is{acc} ")
```
